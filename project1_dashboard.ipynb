```python
#Import necessary libraries
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import pandas as pd
import panel as pn
pn.extension('plotly')
import plotly.express as px
import hvplot.pandas
import os
from pathlib import Path
import cv2
```

```python
#define functions for each portfolio

#Gokhan's portfolio
def gokhan_visuals():

    fig = plt.figure(figsize=(10, 10))

    columns = 3

    rows = 1

    gokhan_img1 = cv2.imread('C:\code/FinTechProject1/Visual Outlook/Gokhan_CRvsBenchmark.png')

    okhan_img2 = cv2.imread('C:\code/FinTechProject1/Visual Outlook/Gokhan_Pie.png')

    gokhan_img3 = cv2.imread('C:\code/FinTechProject1/Visual Outlook/Gokhan_Worst5Drawdown.png')

    #impgplot = plt.imshow(gokhan_img1)

    #plt.show()

    fig.add_subplot(rows, columns, 1)

    plt.imshow(gokhan_img1)
    plt.axis('off')
    plt.title("Gokhan's Portfolio vs Benchmark")

    fig.add_subplot(rows, columns, 2)

    plt.imshow(gokhan_img2)
    plt.axis('off')
    plt.title("Pie Chart")

    fig.add_subplot(rows, columns, 3)

    plt.imshow(gokhan_img3)
    plt.axis('off')
    plt.title("Worst 5 Drawdowns")

    return fig




```

```python
gokhan_tab = pn.Column(gokhan_visuals)
sam_tab = pn.Column(sam_visuals)
william_tab = pn.Column(william_visuals)
```
