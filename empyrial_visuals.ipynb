```python
from empyrial import empyrial, Engine
```

```python
gokhan_portfolio = Engine(
      start_date = "2012-01-01",
      benchmark = ["SPY"], #SPY is set by default
      portfolio = ["AAPL", "TSLA", "AMZN", "GOOG","MSFT"],
      weights = [0.2, 0.3, 0.1, 0.1, 0.3]
      #optimizer = "EF", #optimizer for portfolio optimization: Efficient Frontier
      #rebalance= "1y" #rebalance every year
)

empyrial(gokhan_portfolio)
```
